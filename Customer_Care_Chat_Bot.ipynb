```python
 import json
```

```python
with open('GL_Bot.json') as file:
        corpus=json.load(file)
```

```python
#Display corpus
print(corpus)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer

# Extract data 
# Tokens 
W = []
# Identified Tags or Labels
L = [] 
# Tokenised words
doc_X = [] 
# Tags or Labels
doc_y = [] 

for intent in corpus['intents']:
    for pattern in intent['patterns']:
        w_temp = word_tokenize(pattern) 
        W.extend(w_temp) 
        doc_X.append(w_temp) 
        doc_y.append(intent["tag"])
        # Add the mising tag if any 
        if intent['tag'] not in L:
            L.append(intent['tag'])
            
# Stemming 
# Stemming or learning the root word
stemmer = SnowballStemmer(language='english')
W = [stemmer.stem(w.lower()) for w in W if w != '?']
# Sorted word
W = sorted(list(set(W)))

# Sorted list of tags or labels
L = sorted(L) 

# Training data for NN 
Train = [] 
# Target data for NN
Target = [] 

out_empty=[]
for i in range(len(L)):
    out_empty.append(0)

# Loop to create bag of words and put the frequency count on each word 
for x, doc in enumerate(doc_X):
    
    bag = []    
    w_temp = [stemmer.stem(w.lower()) for w in doc]
    
    for w in W: 
        if w in w_temp:
            bag.append(1)
        else:
            bag.append(0)
            
    output_row = out_empty[:] 
    output_row[L.index(doc_y[x])] = 1
                       
    Train.append(bag)  
    Target.append(output_row) 
```

```python
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, precision_recall_curve, auc
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD

model = Sequential()
model.add(Dense(100, activation='relu', kernel_initializer='he_uniform'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(8, activation='softmax'))
opt = SGD(lr=0.01, momentum=0.9)

model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

history=model.fit(Train, Target, batch_size = 10, epochs=100, verbose=1)
```

```python
import numpy
import random

# Text chat utility function
def chat():
    
    print("Chat with Ramos (type: quit to quit)") 
    print("If answer is not right (type: *)") 
    
    while True:
        
        input_from_user = input("\n\n You: ") 
        
        if input_from_user.lower()=="*":
            print("BOT: Please rephrase your question and try again")
        if input_from_user.lower() == "quit":
            break
        
        #input_from_user='useless piece of shit'

        #print('input from user:',input_from_user)
        #print(' ')

        doc_X.append(word_tokenize(input_from_user))

        # Loop to create bag of words and put the frequency count on each word 
        for x, doc in enumerate(doc_X):

            bag = []    
            w_temp = [stemmer.stem(w.lower()) for w in doc]

            for w in W: 
                if w in w_temp:
                    bag.append(1)
                else:
                    bag.append(0)

        #print('bag of words:',bag)
        #print(' ')

        results=model.predict([bag]) 

        #print('results:',results)
        #print(' ')

        results_index = numpy.argmax(results) 

        tag = L[results_index]

        #print('tag:',tag)
        
        for tg in corpus['intents']:
            if tg['tag']==tag:
                responses=tg['responses']
                
        print(random.choice(responses))          
                  
```
